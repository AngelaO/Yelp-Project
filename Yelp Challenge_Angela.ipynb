```python
def read_mongo(collection, query={}, limit=3):
    """ Read from Mongo and Store into DataFrame """
    # Collection
    cursor = collection.find(query).limit(limit)

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(list(cursor))

    return df
```

```python
# {} - can use as empty query to return all results
# first argument of find is the query criteria, second is the projection i.e. the field
# use "_id":0 so that hash table values aren't returned (they are by default)

# return categories and city for all cities that aren't Carnegie
myresult = businesses.find({"city":{'$ne': "Carnegie"}},{"categories":1, "_id":0, "city":1})
```

```python
myDF = pd.DataFrame(list(myresult))
```
